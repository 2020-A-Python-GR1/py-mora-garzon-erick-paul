```python
class Sistema:
    sistema_id = None
    nombre = None
    version = 0
    fecha_lanzamiento = None
    peso_GB = 0.0
    instalado = False
    aplicaciones = []
    numero_aplicaciones = len(aplicaciones)
    
      
    def __init__(self, sistema_id, nombre, version, fecha_lanzamiento, peso_GB, instalado):
        self.sistema_id = sistema_id
        self.nombre = nombre
        self.version = version
        self.fecha_lanzamiento = fecha_lanzamiento
        self.peso_GB = peso_GB
        self.instalado = instalado
        
    def get_id(self):
        return f"{self.sistema_id}"
    
    def get_nombre(self):
        return f"{self.nombre}"
    
    def set_nombre(self, nombre_SO):
        self.nombre = nombre_SO
        
    def set_version(self, version_SO):
        self.version = version_SO
        
    def set_fecha(self, fecha_lanzamiento):
        self.fecha_lanzamiento = fecha_lanzamiento
        
    def set_peso(self, peso_GB):
        self.peso_GB = peso_GB
        
    def set_instalado(self, instalado):
        self.instalado = instalado
        
    def esta_instalado(self):
        if self.instalado:
            return "Si"
        else:
            return "No"
    
    def nueva_aplicacion(self, nueva_aplicacion):
        self.aplicaciones.append(nueva_aplicacion)
        self.numero_aplicaciones = len(self.aplicaciones)
        return "Aplicacion registrada"
    
    def __str__(self):
        return (f"Datos del Sistema Operativo: \n"+
                f"Id: {self.sistema_id}\n" +
                f"Nombre: {self.nombre}\n" +
                f"Version: {self.version}\n" +
                f"Fecha de lanzamiento: {self.fecha_lanzamiento}\n" +
                f"Peso en GB: {self.peso_GB}\n" +
                f"Numero de aplicaciones: {self.numero_aplicaciones}\n" +
                f"¿Esta instalado? {self.esta_instalado()}\n\n")
```

```python
class Aplicacion:
    aplicacion_id = None
    nombre = None
    version = 0
    precio = 0.0
    peso_GB = 0.0
    instalado = False
    id_sistema = None
    
    def get_id(self):
        return f"{self.aplicacion_id}"
    
    def get_nombre(self):
        return f"{self.nombre}"
    
    def set_nombre(self, nombre_app):
        self.nombre = nombre_app
        
    def set_version(self, version_app):
        self.version = version_app
        
    def set_precio(self, precio):
        self.precio = precio
        
    def set_peso(self, peso_GB):
        self.peso_GB = peso_GB
    
    def set_instalado(self, instalado):
        self.instalado = instalado
        
    def esta_instalado(self):
        if self.instalado:
            return "Si"
        else:
            return "No"
      
    def __init__(self, aplicacion_id, nombre, version, precio, peso_GB, instalado, id_sistema):
        self.aplicacion_id = aplicacion_id
        self.nombre = nombre
        self.version = version
        self.precio = precio
        self.peso_GB = peso_GB
        self.instalado = instalado
        self.id_sistema = id_sistema
    
    def __str__(self):
        return (f"Datos de la aplicacion: \n"+
                f"Id: {self.aplicacion_id}\n" +
                f"Nombre: {self.nombre}\n" +
                f"Version: {self.version}\n" +
                f"Precio: {self.precio}\n" +
                f"Peso en GB: {self.peso_GB}\n" +
                f"¿Esta instalado? {self.esta_instalado()}\n" +
                f"Id SO: {self.id_sistema}\n\n")
```

```python
def leer_archivo(ruta_archivo):
    try:
        archivo_abierto = open(ruta_archivo)
        archivo_completo = archivo_abierto.readlines();  # Devuelve una lista [] con las lineas del archivo
        for linea in archivo_completo:
            print(linea)
        print(f'Archivo completo: {archivo_completo}', type(archivo_completo))
        archivo_abierto.close()
    except Exception as error:
        print('Error al leer archivo')
```

```python
def escribir_archivo(ruta_archivo, informacion, modo):
    try:
        if (modo == "a"):
            archivo_escritura = open(ruta_archivo, mode="a") 
            archivo_escritura.write(informacion)
        elif (modo == "w"):
            archivo_escritura = open(ruta_archivo, mode="w")
            archivo_escritura.writelines(informacion)
        archivo_escritura.close()
    except Exception as error:
        print('Error al escribir archivo')
```

```python
def crear_sistema(id_actual):
    print("Ingresa los datos del sistema operativo:")
    nombre = input("Nombre: ")
    version = input("Version: ")
    fecha_lanzamiento = input("Fecha de lanzamiento: ")
    peso_GB = input("Peso en GB: ")
    instalado = input("Esta instalado? si/no: ")
    if(instalado == "si" or instalado == "Si" or instalado == "SI"):
        instalado = True
    else:
        instalado = False
    nuevo_sistema = Sistema(id_actual,nombre,version,fecha_lanzamiento,peso_GB,instalado)
    return nuevo_sistema
```

```python
def menu_aplicaciones(sistema, sistema_id, id_actual_app):
    opcion = None
    aplicaciones = sistema.aplicaciones
    #id_actual = 1
    while(opcion != 0):
        print("\n")
        print(f"Gestion de aplicaciones del sistema operativo {sistema.nombre}")
        print("\n")
        print("1 - Crear nueva aplicacion")
        print("2 - Buscar aplicacion")
        print("3 - Listar aplicaciones")
        print("4 - Eliminar aplicacion")
        print("5 - Editar aplicacion")
        print("0 - Regresar")
        print("\n")
        print("Escoge una opcion del menu:")
        opcion = int(input("Opcion:"))
        print("\n")
        if(opcion == 1):
            print("---Creacion de aplicacion---")
            nueva_app = crear_aplicacion(id_actual_app, sistema_id)
            sistema.nueva_aplicacion(nueva_app)
            #aplicaciones.append(nueva_app)
            try:
                escribir_archivo("./aplicaciones.txt",f"{nueva_app}","a")
            except Exception as error:
                print("Ocurrio un error al escribir archivo")
            id_actual_app = id_actual_app + 1
            print("Aplicacion registrada")
        elif(opcion == 2):
            print("---Buscar aplicacion---")
            buscar_aplicacion(aplicaciones)
        elif(opcion == 3):
            print("---Aplicaciones registradas---")
            if(len(aplicaciones) > 0):
                for aplicacion in aplicaciones:
                    print(aplicacion)
            else:
                print("No hay aplicaciones registradas")
        elif(opcion == 4):
            print("---Eliminar aplicacion---")
            print("\n")
            aplicacion_id = input("Ingrese el id de la aplicacion a eliminar: ")
            aplicacion = buscar_aplicacion_por_id(aplicacion_id, aplicaciones)
            if (aplicacion):
                del aplicaciones[int(aplicacion.get_id()) - 1]                
                try:
                    escribir_archivo("./aplicaciones.txt",f"","w")
                    if(len(aplicaciones) > 0):
                        for aplicacion in aplicaciones:
                            escribir_archivo("./aplicaciones.txt",f"{aplicacion}","a")
                    else:
                        print("No hay sistemas registrados")  
                except Exception as error:
                    print("Ocurrio un error al escribir archivo")
                print("Aplicacion eliminada")
            else:
                print("Id no encontrado")
        elif(opcion == 5):
            print("---Editar aplicacion---")
            print("\n")
            aplicacion_id = input("Ingrese el id de la aplicacion a editar: ")
            aplicacion = buscar_aplicacion_por_id(aplicacion_id, aplicaciones)
            if (aplicacion):
                editar_aplicacion(aplicacion)
                try:
                    escribir_archivo("./aplicaciones.txt",f"","w")
                    if(len(aplicaciones) > 0):
                        for aplicacion in aplicaciones:
                            escribir_archivo("./aplicaciones.txt",f"{aplicacion}","a")
                    else:
                        print("No hay sistemas registrados")                    
                except Exception as error:
                    print("Ocurrio un error al escribir archivo")
                print("Aplicacion actualizada")
            else:
                print("Id no encontrado")
        elif(opcion == 0):
            print("Menu ...")
        else:
            print("Opcion no valida")
            opcion = 0
```

```python
def main():
    sistemas = []
    opcion = None
    id_actual = 1
    id_actual_app = 1
    while(opcion != 0):
        print("\n")
        print("Gestion de sistemas operativos")
        print("\n")
        print("1 - Crear nuevo sistema")
        print("2 - Editar sistema")
        print("3 - Buscar sistema")
        print("4 - Listar sistemas")
        print("5 - Eliminar sistema")
        print("6 - Gestionar aplicaciones")
        print("0 - Salir")
        print("\n")
        print("Escoge una opcion:")
        print("\n")
        opcion = int(input("Opcion: "))
        print("\n")
        if(opcion == 1):
            print("---Creacion de sistema---")
            nuevo = crear_sistema(id_actual)
            sistemas.append(nuevo)
            try:
                escribir_archivo("./sistemas.txt",f"{nuevo}","a")
            except Exception as error:
                print("Ocurrio un error al escribir archivo")
            id_actual = id_actual + 1
            print("Sistema registrado con exito")
        elif(opcion == 2):
            print("---Editar sistema---")
            sistema_id = input("Ingrese el id del sistema a editar: ")
            sistema = buscar_sistema_por_id(sistema_id, sistemas)
            if (sistema):
                editar_sistema(sistema, sistema_id)                
                try:
                    escribir_archivo("./sistemas.txt",f"","w")
                    if(len(sistemas) > 0):
                        for sistema in sistemas:
                            escribir_archivo("./sistemas.txt",f"{sistema}","a")
                    else:
                        print("No hay sistemas registrados")                    
                except Exception as error:
                    print("Ocurrio un error al escribir archivo")
                print("Sistema actualizado")
            else:
                print("Id no encontrado")
        elif(opcion == 3):
            print("---Buscar sistema---")
            buscar_sistema(sistemas)
        elif(opcion == 4):
            print("---Sistemas registrados---")
            print("\n")
            if(len(sistemas) > 0):
                for sistema in sistemas:
                    print(sistema)
            else:
                print("No hay sistemas registrados")
        elif(opcion == 5):
            print("---Eliminar sistema---")
            sistema_id = input("Ingrese el id del sistema a eliminar: ")
            sistema = buscar_sistema_por_id(sistema_id, sistemas)
            if (sistema):
                del sistemas[int(sistema.get_id()) - 1]
                try:
                    escribir_archivo("./sistemas.txt",f"","w")
                    if(len(sistemas) > 0):
                        for sistema in sistemas:
                            escribir_archivo("./sistemas.txt",f"{sistema}","a")
                    else:
                        print("No hay sistemas registrados")  
                except Exception as error:
                    print("Ocurrio un error al escribir archivo")
                print("Sistema eliminado")
            else:
                print("Id no encontrado")
        elif(opcion == 6):
            print("---Gestión de aplicaciones---")
            print("\n")
            sistema_id = input("Ingrese el id del sistema: ")
            sistema = buscar_sistema_por_id(sistema_id, sistemas)
            if (sistema):
                menu_aplicaciones(sistema, sistema_id, id_actual_app)
            else:
                print("Id no encontrado")
        elif(opcion == 0):
            print(" ")
        else:
            print("Opcion no valida")
            opcion = 0
```
